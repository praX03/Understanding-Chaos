```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
%matplotlib notebook
N = 500
x = 0.5+np.zeros(N)
lambda_c = np.linspace(1, 4, 1000)
endcap = np.arange(round(N*0.9), N)

for ri in range(len(lambda_c)):
    for n in range(N-1):
        x[n+1] = lambda_c[ri]*x[n]*(1 - x[n-1])
    u = np.unique(x[endcap])
    r = lambda_c[ri]*np.ones(len(u))
    plt.plot(r, u, 'k.')
plt.show()
```
